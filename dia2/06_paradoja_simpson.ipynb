# PARADOJA DE SIMPSONS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins_df = sns.load_dataset('penguins')
penguins_df
```

## ELIMINAMOS NULOS

```python
penguins_df.dropna(inplace=True)
penguins_df
```

## CORRELACIÃ“N DE VARIABLES

```python
sns.clustermap(
    data=penguins_df.corr(numeric_only=True),
    cmap=sns.diverging_palette(20,230,as_cmap=True),
    center=0,
    vmin=-1,
    vmax=1,
    linewidths=0.5,
    cbar_kws={'shrink':0.5},
    annot=True
)
```

# PARADOJA DE SIMPSON

## ANALIZAMOS LA RELACIÃ“N TOTAL ENTRE EL BILL_LENGTH_MM y EL BILL_DEPTH_MM

```python

%matplotlib inline
sns.set_style(style='whitegrid')
sns.set_context(context='notebook')
plt.rcParams['figure.figsize'] = (11, 9.4)

penguin_color = {
    'Adelie': '#ff6602ff',
    'Gentoo': '#0f7175ff',
    'Chinstrap': '#c65dc9ff'
}
```

```python
sns.scatterplot(
    data=penguins_df,
    x='bill_length_mm',
    y='bill_depth_mm'
)
```

## ANALIZAMOS RELACIÃ“N PARCIAL (AGRUPAR LOS RESULTADOS POR LA SPECIE)

```python
sns.scatterplot(
    data=penguins_df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    palette=penguin_color
)
```

## GRAFICAMOS UNA REGRESION LINEAL SIMPLE PARA ESTA AGRUPACIÃ“N

```python
sns.lmplot(
    data=penguins_df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    height=10,
    palette=penguin_color
)
```

## APLICAMOS LA AGRUPACIÃ“N PARA EL FLIPPER_LENGTH_MM Y BODY_MASS_G

```python
sns.lmplot(
    data=penguins_df,
    x='flipper_length_mm',
    y='body_mass_g',
    hue='species',
    height=10,
    palette=penguin_color
)
```

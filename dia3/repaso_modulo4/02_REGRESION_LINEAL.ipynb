# CORRELACIONES

```python
sns.clustermap(
    data=cars_df.corr(numeric_only=True),
    cmap=sns.diverging_palette(20, 230, as_cmap=True),
    center=0,
    vmin=-1,
    vmax=1,
    linewidths=0.5,
    cbar_kws={"shrink": 0.5},
    annot=True
)
```

# REGRESIÓN LINEAL SIMPLE

```python
model_1 = (
    smf.ols(
        formula = 'Price ~ Year',
        data=cars_df
    ).fit()
)
model_1.summary()
```

# REGRESIÓN LINEAL MULTIPLE

```python
model_2 = (
    smf.ols(
        formula = 'Price ~ Year + Engine_Size + Mileage',
        data=cars_df
    ).fit()
)
model_2.summary()
```

```python
model_3 = (
    smf.ols(
        formula = 'Price ~ Year + Engine_Size + Mileage + C(Brand) + C(Fuel_Type) + C(Transmission)',
        data=cars_df
    ).fit()
)
model_3.summary()
```

```python
summary_table = model_3.summary2().tables[1]
sorted_values = summary_table.sort_values(by='P>|t|',ascending=False)
sorted_values
```

```python
model_4 = (
    smf.ols(
        formula = 'Price ~ Year + Engine_Size + Mileage + C(Fuel_Type) + C(Transmission)',
        data=cars_df
    ).fit()
)
model_4.summary()
```

```python
model_5 = (
    smf.ols(
        formula = 'Price ~ Year + Engine_Size + Mileage + C(Transmission)',
        data=cars_df
    ).fit()
)
model_5.summary()
```

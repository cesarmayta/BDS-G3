# PROBANDO CON NUESTRO DATASET DUMMY

```python
df = pd.DataFrame.from_dict(
    data = {
        "a": list('asdfasdfas'),
        "b": range(0,10)
    }
)
df.iloc[2:5,0] = None
df.iloc[6:7,1] = None
df
```

```python
print(f"total de valores: {df.size}")
print(f"total de valores missing: {df.missing.number_missing()}")
print(f"total de valores completos: {df.missing.number_complete()}")
```

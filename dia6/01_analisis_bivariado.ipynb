# PASO 1 - IMPORTACIÓN DE LIBRERIAS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# PASO 2 - ESTILOS DE GRAFICOS

```python
%matplotlib inline
sns.set_style(style='whitegrid')
sns.set_context(context='notebook')
plt.rcParams['figure.figsize'] = (11, 9.4)

penguin_color = {
    'Adelie': '#ff6602ff',
    'Gentoo': '#0f7175ff',
    'Chinstrap': '#c65dc9ff'
}
```

# PASO 3 - CARGA DE DATASET

```python
penguins_df = sns.load_dataset('penguins')
penguins_df
```

# ANALIS BIVARIADO

## RELACIONES ENTRE 2 VARIABLES NÚMERICAS

```python
sns.scatterplot(
    data=penguins_df,
    x='bill_length_mm',
    y='bill_depth_mm',
    alpha=1/2,
    s=100
)
```

```python
sns.scatterplot(
    data=penguins_df,
    x='flipper_length_mm',
    y='body_mass_g',
    alpha=1/2,
    s=100
)
```

```python
sns.scatterplot(
    data=penguins_df,
    x='bill_depth_mm',
    y='body_mass_g',
    alpha=1/2,
    s=100
)
```

```python
sns.scatterplot(
    data=penguins_df,
    x='bill_length_mm',
    y='body_mass_g',
    alpha=1/2,
    s=100
)
```

## RELACIÓN ENTRE VARIABLE NÚMERICA Y CATEGORICA

### RELACIÓN ENTRE FLIPPER_LENGTH_MM Y SPECIES

```python
sns.scatterplot(
    data=penguins_df,
    x='species',
    y='flipper_length_mm',
    hue='species',
    palette=penguin_color
)
```

### GENERAMOS RUIDO EN LOS VALORES PARA UNA MEJOR VISUALIZACIÓN

```python
sns.stripplot(
    data=penguins_df,
    x='species',
    y='flipper_length_mm',
    hue='species',
    palette=penguin_color
)
```

### UTILIZAMOS BOXPLOT PARA VER EL RANGO

```python
sns.boxplot(
    data=penguins_df,
    x='flipper_length_mm',
    y='species',
    palette=penguin_color,
    whis=np.inf
)
sns.stripplot(
    data=penguins_df,
    x='flipper_length_mm',
    y='species',
    color='.3'
)
```

### USAMOS VIOLINPLOT PARA TENER UN GRAFICO MAS EXACTO

```python
sns.violinplot(
    data=penguins_df,
    x='species',
    y='flipper_length_mm',
    color='.8'
)

sns.stripplot(
    data=penguins_df,
    x='species',
    y='flipper_length_mm',
    hue='species',
    palette=penguin_color
)
```

### USAMOS SWARMPLOT PARA UNA DISTRIBUCIÓN MAS UNIFORME

```python
sns.swarmplot(
    data=penguins_df,
    x='species',
    y='flipper_length_mm',
    hue='species',
    palette=penguin_color
)
```

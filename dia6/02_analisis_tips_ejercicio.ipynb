# EJERCICIO ANALISIS DE TOTAL_BILL Y TIP Y TIP CON DAY USANDO TODO LO APRENDIDO

## RELACIÓN ENTRE TOTAL_BILL Y TIP

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.scatterplot(
    data=tips_df,
    x='total_bill',
    y='tip',
    alpha=1/2,
    s=100
)
```

```python
sns.displot(
    data=tips_df,
    x='total_bill',
    y='tip',
    rug=True
)
```

```python
sns.displot(
    data=tips_df,
    x='total_bill',
    y='tip',
    rug=True,
    kind='kde'
)
```

## RELACIÓN DEL TIP CON EL DAY

```python
sns.stripplot(
    data=tips_df,
    x='day',
    y='tip',
    hue='day'
)
```

```python
sns.boxplot(
    data=tips_df,
    x='day',
    y='tip',
    whis=np.inf
)
sns.stripplot(
    data=tips_df,
    x='day',
    y='tip',
    color='.3'
)
```

```python
sns.violinplot(
    data=tips_df,
    x='day',
    y='tip',
    color='.8'
)

sns.stripplot(
    data=tips_df,
    x='day',
    y='tip',
    hue='day'
)
```

```python
sns.swarmplot(
    data=tips_df,
    x='day',
    y='tip',
    hue='day'
)
```
